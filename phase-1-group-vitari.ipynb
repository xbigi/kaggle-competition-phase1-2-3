```python
# importo le librerie che ci servono
import pandas as pd  # pandas serve per lavorare con tabelle di dati
from sklearn.ensemble import RandomForestClassifier  # questa libreria sarà per il nostro modello
from sklearn.preprocessing import LabelEncoder  # per convertire testo in numeri perché i modelli di machine learning (come RandomForestClassifier) lavorano solo con numeri



# carico i dati da file (train e test)
train = pd.read_csv('/kaggle/input/phase-1-competition/train.csv')
test = pd.read_csv('/kaggle/input/phase-1-competition/test.csv')


#  X contiene le colonne che useremo per prevedere
#  y è la colonna che vogliamo prevedere (lo "status")
X = train.drop(['id', 'Status'], axis=1)
y = train['Status']

#  convertovlo "status" (che è in lettere C, CL, D) in numeri (0, 1, 2)
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

#  convertiamo le colonne con testo tipo 'sex', 'drug' e le altre in numeri usando One-Hot Encoding
# questo crea nuove colonne per ogni valore possibile
X = pd.get_dummies(X)
X_test = pd.get_dummies(test.drop(['id'], axis=1))


X, X_test = X.align(X_test, join='left', axis=1, fill_value=0)

# 6. creo il modello random forest (preso dalla guida base)
model = RandomForestClassifier()
model.fit(X, y)  # alleno il modello con i dati del file train.csv

#  faccio le previsioni  il modello ci dice le probabilità per ogni classe
probs = model.predict_proba(X_test)

# creo il file da consegnare a kaggle
# ogni riga contiene l'id del paziente e le probabilità per C, CL e D
submission = pd.DataFrame({
    'id': test['id'],
    'Status_C': probs[:, 0],
    'Status_CL': probs[:, 1],
    'Status_D': probs[:, 2],
})

# salvo il file dell output in formato csv che è un formato excel che divide i dati con i punti e virgola
submission.to_csv('/kaggle/working/submission.csv', index=False)
```
